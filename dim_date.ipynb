```python
import pandas as pd
from datetime import datetime, timedelta

# Create parameter
today = datetime.now().replace(day=31, month=12).date()
start_date = datetime(2018, 1, 1).date()
step = (today - start_date).days
```

```python
# Create list of date
source = pd.date_range(start=start_date, periods=step + 1, freq='d')

# Create dataframe
df = pd.DataFrame(source, columns=['date'])
```

```python
# Create additional columns
df['date_key'] = df['date'].dt.strftime('%y%M%d')
df['year'] = df['date'].dt.year
df['quarter'] = 'q' + df['date'].dt.quarter.astype(str)
df['month'] = df['date'].dt.month
df['month_name'] = df['date'].dt.strftime('%b')
df['week_of_year'] = df['date'].dt.isocalendar().week
df['day_of_week'] = df['date'].dt.dayofweek
df['day_name'] = df['date'].dt.strftime('%a')
```

```python
# reorder columns
df = df[['date_key', 'date', 'year', 'quarter', 'month', 'month_name', 'week_of_year', 'day_of_week', 'day_name']]
df
```

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("Pandas to Spark").getOrCreate()
dim_date_df = spark.createDataFrame(df)
dim_date_df.printSchema()
```

```python
dim_date_df.write.format('delta').mode('append').option("delta.columnMapping.mode", "name").save('Tables/dim_date')
```
